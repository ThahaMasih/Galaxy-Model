VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

-HISTPLOT

```python
sns.histplot(df.median_income,color="blue",alpha=0.4,kde=True)
plt.show()
```

-HEATMAP

```python
k=df.corr()
sns.heatmap(k,annot=True,cmap='bwr')
```

-KDEPLOT

```python
sns.kdeplot([df.households,df.population],palette=['green','blue'],alpha=0.7)
```
