```python
from astropy.modeling.models import Gaussian2D
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = np.mgrid[0:200,0:200]
x
```

```python
model = Gaussian2D(1,100,100,20,20,np.pi/4)
model
```

```python
galaxy_model = model(x,y)
plt.imshow(galaxy_model,cmap='jet')
```
